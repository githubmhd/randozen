```python
url = "https://www.gironde-tourisme.com/sur-place/balades/la-gironde-a-pied/les-boucles-de-randonnee"
suite = ["/", "/page/2", "/page/3"]
difficulty = []
duree = []
distance = []
ville = []
name = []
types = []
image = []


for pages in suite:
    
    page = requests.get(url + pages)
    soup = BeautifulSoup(page.text, "html.parser")
    rando = soup.find_all("div", {"class": "wpetOfferContent"})
    pic = soup.find_all("div", {"class": "wpetOfferContainer"})
    
    for line in rando:
        try:
            difficulty.append(line.find("div", class_ = "wpetOfferContainerContentItinerarySubItem wpetOfferContainerContentItinerarySubItem--difficulte").text)
        except:
            difficulty.append("")
            
    for line in rando:
        try:
            duree.append(line.find("div", class_ = "wpetOfferContainerContentItinerarySubItem wpetOfferContainerContentItinerarySubItem--duree").text)
        except:
            duree.append("")

    for line in rando:
        distance.append(line.find("span", {"class": "wpetOfferContainerContentItineraryItem wpetOfferContainerContentItineraryItem--distance"}).text)
        
    for line in rando:
        try:
            types.append(line.find("div", class_ = "wpetOfferContainerContentItinerarySubItem").text)
        except:
            types.append("")
    
    for line in rando:
        ville.append(line.find("span", {"class": "wpetOfferContentMetaLocation"}).text)
        
        
    for line in rando:
        name.append(line.find("h3", {"class": "wpetOfferContentTitle"}).text)
        
    for ele in pic:
        try:
            image.append(ele.find("img")["src"])
        except:
            image.append("")
        
```
